```python
import pandas as pd
```

```python
#read the csv file
df=pd.read_csv('ETH_1h.csv')
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'], format='%Y-%m-%d %I-%p')
```

```python
df['Date']
```

```python
df.loc[0,'Date'].day_name()
```

```python
df['Date'].dt.day_name()
```

```python
#create a new column to mark the day of the week
df['dayOfweek']=df['Date'].dt.day_name()
df
```

```python
#gives the earliest date in our dataframe
df['Date'].min()
```

```python
#gives the most recent date 
df['Date'].max()
```

```python
#deltatime shows the days between the earliest and the last date
df['Date'].max()-df['Date'].min()
```

```python
filt=(df['Date'] >'2019') & (df['Date'] <'2020')
df.loc[filt]
```

```python
#set the Date column as an index
df.set_index('Date', inplace=True)
df
```

```python
df['2019':'2020']
```

```python
#get all the High value in the periode between 2019 and 2020
df['2019':'2020']['High']
```

```python
#resample is a Convenience method for frequency conversion and resampling of time series, 
#in our case we use  to resample into DAYS
#all the High values for each Day 'D'
highs=df['High'].resample('D').max()
highs
```

```python
#let's install the matplotlib in our virtual environement to display some plots directly 
%matplotlib inline
```

```python
highs.plot()
```

```python
#we want to calculate the avg of each column and resample our dataframe by week 'W'
Avg=df.resample('W').mean()
```

```python
Avg[['Open', 'High']].plot()
```

```python
#here we pass an aggregation function for eanch colums and resample that by week 
df.resample('W').agg({'Close':'mean', 'High':'max', 'Low':'min', 'Volume':'sum'})
```
