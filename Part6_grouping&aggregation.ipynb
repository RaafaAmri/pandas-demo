```python
#here we want to group all the result by country
country_grp=df.groupby(['Country'])
```

```python
#in this case the country we want to group by is "United States"
country_grp.get_group('United States')
```

```python
#display the first 15 lines of column "SocialMedia" grouped by country
country_grp['SocialMedia'].value_counts().head(15)
```

```python
#in this case we specify the country(united states) that we want to grouped by
country_grp['SocialMedia'].value_counts().loc['United States']
```

```python
#the median slary of all the countries
country_grp['ConvertedComp'].median()
```

```python
#in case we want to do some analysis and we want to run a multiple of aggregation function 
country_grp['ConvertedComp'].agg(['median', 'mean'])
```

```python
#we can specify the country by using loc['NameOfCountry']
country_grp['ConvertedComp'].agg(['median', 'mean']).loc['Canada']
```

```python
#all the people from each country who use python
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```
