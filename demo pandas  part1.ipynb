```python
import pandas as pd
pd.set_option('display.max_columns', None)  # 'None' mean that we want to display all rows and columns, 
pd.set_option('display.max_rows', None) #we can specify the number we want to display
```

```python
#read the csv file and specify the id as an index
df=pd.read_csv('MOCK_DATA.csv',index_col='id')
df
```

```python
#return the columns in our dataframe
df.columns
```

```python
#by default the head fuction display the first 5 rows, and we can specify the number into the brackets 
df.head(10)
```

```python
#by default display the last 5 rows, and we can specify the number into the brackets 
df.tail(2)
```

```python
#return the dimension of the DataFrame
df.shape
```

```python
#display a specific column exemple 'last_name'
df['last_name']
```

```python
#loc is used to specify both row and column with label
#in this exemple we want to display the row number 2 
df.loc[2]
```

```python
#display rows from row 1 to row 10 
df.loc[1:10]
```

```python
#also we can specifiy wich column we want to display, in this exemple we want to display cols from emailt to ip_address
df.loc[1:10,'email':'ip_address']
```

```python
#to access multiple lines use double brackets separated by commas and specify the labels
df [['email', 'gender']]
df.head(15)
```

```python
#iloc ollows to access row by location (position)
df.iloc[1]
```

```python
df.iloc[[0,1]]
```

```python
#we can separate rows and coloumns with the double brakets, the first brakets[0,1] means rows with index 0 and 1, 
#then the column with index 1 wich is 'last name'
df.iloc[[0,1],1]
```

```python
#here we can get the same result by using loc[] with labels 
df.loc[0:2,'last_name']
```
